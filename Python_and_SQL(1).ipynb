```python
query = '''
SELECT
    ROUND(AVG(salary_from)) as avg_salary_from,
    ROUND(AVG(salary_to)) as avg_salary_to
FROM public.vacancies
WHERE salary_from IS NOT NULL
   OR salary_to IS NOT NULL
'''

df = pd.read_sql_query(query, connection)
df
```

```python
query = '''
WITH schedule_employment_counts AS (
    SELECT
        schedule,
        employment,
        COUNT(*) as vacancy_count,
        ROW_NUMBER() OVER (ORDER BY COUNT(*) DESC) as rank
    FROM public.vacancies
    GROUP BY schedule, employment
)
SELECT
    schedule,
    employment,
    vacancy_count
FROM schedule_employment_counts
WHERE rank = 2
'''

df = pd.read_sql_query(query, connection)
df
```

```python
query = '''
SELECT
    experience,
    COUNT(*) as vacancy_count
FROM public.vacancies
GROUP BY experience
ORDER BY vacancy_count ASC
'''

df = pd.read_sql_query(query, connection)
df
```

```python
query = '''
WITH employer_vacancy_rank AS (
    SELECT
        e.name as employer_name,
        COUNT(v.id) as vacancy_count,
        ROW_NUMBER() OVER (ORDER BY COUNT(v.id) DESC) as rank
    FROM public.employers e
    JOIN public.vacancies v ON e.id = v.employer_id
    GROUP BY e.id, e.name
)
SELECT
    employer_name,
    vacancy_count,
    rank
FROM employer_vacancy_rank
WHERE rank IN (1, 5)
ORDER BY rank
'''

df = pd.read_sql_query(query, connection)
df
```

```python
query = '''
WITH region_stats AS (
    SELECT
        a.id,
        a.name as region_name,
        COUNT(DISTINCT e.id) as employers_count,
        COUNT(DISTINCT v.id) as vacancies_count
    FROM public.areas a
    LEFT JOIN public.employers e ON a.id = e.area
    LEFT JOIN public.vacancies v ON a.id = v.area_id
    GROUP BY a.id, a.name
)
SELECT
    region_name,
    employers_count,
    vacancies_count
FROM region_stats
WHERE vacancies_count = 0
ORDER BY employers_count DESC
LIMIT 1
'''

df = pd.read_sql_query(query, connection)
df
```

```python
query = '''
SELECT
    e.name as employer_name,
    COUNT(DISTINCT v.area_id) as regions_count
FROM public.employers e
JOIN public.vacancies v ON e.id = v.employer_id
GROUP BY e.id, e.name
ORDER BY regions_count DESC
LIMIT 1
'''

df = pd.read_sql_query(query, connection)
df
```

```python
query = '''
SELECT COUNT(*) as employers_without_industry
FROM public.employers e
LEFT JOIN public.employers_industries ei ON e.id = ei.employer_id
WHERE ei.industry_id IS NULL
'''

df = pd.read_sql_query(query, connection)
df
```

```python
query = '''
WITH companies_with_4_industries AS (
    SELECT
        e.id,
        e.name as company_name,
        COUNT(ei.industry_id) as industries_count
    FROM public.employers e
    JOIN public.employers_industries ei ON e.id = ei.employer_id
    GROUP BY e.id, e.name
    HAVING COUNT(ei.industry_id) = 4
)
SELECT
    company_name,
    industries_count
FROM companies_with_4_industries
ORDER BY company_name ASC
LIMIT 1 OFFSET 2
'''

df = pd.read_sql_query(query, connection)
df
```

```python
query = '''
WITH yandex_vacancies AS (
    SELECT
        a.name as city,
        COUNT(v.id) as vacancy_count
    FROM public.vacancies v
    JOIN public.employers e ON v.employer_id = e.id
    JOIN public.areas a ON v.area_id = a.id
    WHERE e.name = 'Яндекс'
    AND a.name IN ('Москва', 'Санкт-Петербург', 'Новосибирск', 'Екатеринбург',
                   'Нижний Новгород', 'Казань', 'Челябинск', 'Омск',
                   'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск',
                   'Пермь', 'Воронеж', 'Волгоград', 'Краснодар')  -- Добавлен Краснодар
    GROUP BY a.name
),
total_count AS (
    SELECT SUM(vacancy_count) as total_vacancies
    FROM yandex_vacancies
)
SELECT
    city,
    vacancy_count
FROM yandex_vacancies
UNION ALL
SELECT
    'Total' as city,
    total_vacancies as vacancy_count
FROM total_count
ORDER BY vacancy_count ASC, city ASC
'''

df = pd.read_sql_query(query, connection)
df
```

```python
query = '''
SELECT
    AVG(
        CASE
            WHEN key_skills IS NULL OR key_skills = '' THEN 0
            ELSE LENGTH(key_skills) - LENGTH(REPLACE(key_skills, CHR(10), '')) + 1
        END
    )::numeric as avg_skills
FROM public.vacancies
WHERE (
    LOWER(name) LIKE '%data%scientist%'
    OR LOWER(name) LIKE '%data%science%'
    OR LOWER(name) LIKE '%исследователь%данн%'
    OR (LOWER(name) LIKE '%ml%' AND LOWER(name) NOT LIKE '%html%')
    OR LOWER(name) LIKE '%machine%learning%'
    OR LOWER(name) LIKE '%машинн%обучен%'
)
'''

df = pd.read_sql_query(query, connection)
# Форматируем результат с двумя знаками после запятой
formatted_result = f"{df['avg_skills'][0]:.2f}"
print(f"Среднее количество навыков: {formatted_result}")
```

```python
query = '''
SELECT
    experience,
    ROUND(
        AVG(
            COALESCE(salary_to, salary_from) +
            COALESCE(salary_from, salary_to)
        ) / 2
    ) as avg_salary
FROM public.vacancies
WHERE (
    -- Условия для data science вакансий
    LOWER(name) LIKE '%data%scientist%'
    OR LOWER(name) LIKE '%data%science%'
    OR LOWER(name) LIKE '%исследователь%данн%'
    OR (LOWER(name) LIKE '%ml%' AND LOWER(name) NOT LIKE '%html%')
    OR LOWER(name) LIKE '%machine%learning%'
    OR LOWER(name) LIKE '%машинн%обучен%'
)
AND (
    -- Только вакансии с указанной зарплатой
    salary_from IS NOT NULL
    OR salary_to IS NOT NULL
)
AND experience = 'От 3 до 6 лет'
GROUP BY experience
'''

df = pd.read_sql_query(query, connection)
df
```
